```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from PIL import Image
import numpy as np
import pywt
import matplotlib.pyplot as plt

# ===== Step 1: Define GENIUS-Net Components =====

class FrequencyEncoder(nn.Module):
    def __init__(self):
        super(FrequencyEncoder, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(12, 32, 3, padding=1),  # 4 DWT bands  3 channels
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((16, 16))
        )

    def forward(self, x):
        return self.model(x)

class SpatialEncoder(nn.Module):
    def __init__(self):
        super(SpatialEncoder, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((16, 16))
        )

    def forward(self, x):
        return self.model(x)

class GENIUSNet(nn.Module):
    def __init__(self):
        super(GENIUSNet, self).__init__()
        self.freq_encoder = FrequencyEncoder()
        self.spatial_encoder = SpatialEncoder()
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 16 * 16 * 2, 128),
            nn.ReLU(),
            nn.Linear(128, 2)  # Output: [Real, Fake]
        )

    def forward(self, freq_input, spatial_input):
        f = self.freq_encoder(freq_input)
        s = self.spatial_encoder(spatial_input)
        combined = torch.cat([f, s], dim=1)
        return self.classifier(combined)

# ===== Step 2: Preprocessing Functions =====

def apply_dwt(image_tensor):
    dwt_channels = []
    for channel in image_tensor.squeeze(0):
        coeffs2 = pywt.dwt2(channel.numpy(), 'haar')
        LL, (LH, HL, HH) = coeffs2
        dwt_channels.extend([LL, LH, HL, HH])
    dwt_np = np.stack(dwt_channels)
    dwt_tensor = torch.tensor(dwt_np, dtype=torch.float32).unsqueeze(0)
    return dwt_tensor

def load_image(image_path):
    image = Image.open(image_path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor()
    ])
    img_tensor = transform(image).unsqueeze(0)
    freq_tensor = apply_dwt(img_tensor)
    return image, img_tensor, freq_tensor

# ===== Step 3: Load Image and Predict =====

def predict(image_path):
    image, spatial_input, freq_input = load_image(image_path)
    model = GENIUSNet()
    model.eval()

    with torch.no_grad():
        output = model(freq_input, spatial_input)
        probs = torch.softmax(output, dim=1)
        pred = torch.argmax(probs, dim=1).item()
        label = "AI-Generated" if pred == 1 else "Real"

    # ===== Step 4: Display Image and Prediction =====
    plt.imshow(image)
    plt.axis('off')
    plt.title(f"Prediction: {label} (Confidence: {probs[0][pred]:.2f})")
    plt.show()

# ===== Step 5: Run on Your Image =====
predict("/content/2.gif")  # Replace with your actual image path
```

```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import pywt

# ========== 1. Define GENIUS-Net Components ==========

class FrequencyEncoder(nn.Module):
    def __init__(self):
        super(FrequencyEncoder, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(12, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((16, 16))
        )

    def forward(self, x):
        out = self.model(x)
        print(" Frequency Features Shape:", out.shape)
        return out

class SpatialEncoder(nn.Module):
    def __init__(self):
        super(SpatialEncoder, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((16, 16))
        )

    def forward(self, x):
        out = self.model(x)
        print(" Spatial Features Shape:", out.shape)
        return out

class GENIUSNet(nn.Module):
    def __init__(self):
        super(GENIUSNet, self).__init__()
        self.freq_encoder = FrequencyEncoder()
        self.spatial_encoder = SpatialEncoder()
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 16 * 16 * 2, 128),
            nn.ReLU(),
            nn.Linear(128, 2)
        )

    def forward(self, freq_input, spatial_input):
        freq_feat = self.freq_encoder(freq_input)
        spatial_feat = self.spatial_encoder(spatial_input)
        combined = torch.cat([freq_feat, spatial_feat], dim=1)
        print(" Fused Feature Shape:", combined.shape)
        output = self.classifier(combined)
        print(" Logits Output:", output)
        return output

# ========== 2. DWT Processing ==========

def apply_dwt(image_tensor):
    dwt_channels = []
    for channel in image_tensor.squeeze(0):
        coeffs2 = pywt.dwt2(channel.numpy(), 'haar')
        LL, (LH, HL, HH) = coeffs2
        dwt_channels.extend([LL, LH, HL, HH])
    dwt_np = np.stack(dwt_channels)
    print(" DWT Shape:", dwt_np.shape)
    return torch.tensor(dwt_np, dtype=torch.float32).unsqueeze(0)

# ========== 3. Load and Preprocess Image ==========

def load_image(image_path):
    image = Image.open(image_path).convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor()
    ])
    img_tensor = transform(image).unsqueeze(0)
    print(" Spatial Input Shape:", img_tensor.shape)
    freq_tensor = apply_dwt(img_tensor)
    return image, img_tensor, freq_tensor

# ========== 4. Prediction and Visualization ==========

def predict(image_path):
    image, spatial_input, freq_input = load_image(image_path)
    model = GENIUSNet()
    model.eval()

    with torch.no_grad():
        output = model(freq_input, spatial_input)
        probs = torch.softmax(output, dim=1)
        pred = torch.argmax(probs, dim=1).item()
        label = "AI-Generated" if pred == 1 else "Real"

    #  Display Image with Prediction
    plt.imshow(image)
    plt.title(f"Prediction: {label} (Confidence: {probs[0][pred]:.2f})")
    plt.axis('off')
    plt.show()

# ========== 5. Run Prediction ==========

predict("/content/2.gif")  # Replace with path to your image
```
